<h3> Rules: </h3>
<ul>
    <li> <b>Travel in a loop</b>: The salesman travels in a loop. A loop of length X refers to a tour of X cities, where the salesman returns to the original city after visiting the last (a closed loop), to restock. <br>
    <b>Example</b>: for a loop of length four, the itinerary might look as follows [1, 2, 3, 4]. But, in reality, the class calculates the length with return to 1. <br>
    <li> <b>No revisits</b>: Our salesman only visits each city once in a loop - <b>NOT [1, 2, 1, 2, 1]</b>! The class does not check for this itself, however. You have to make sure that this is obeyed.

<h3> Functions: </h3>
<ul>
    <li> initializing the class creates a world with a given number of cities and resets the random seed. <br>
    <li> random_tour creates a random tour in the world with a given number of stops. <br>
    <li> new_tour creates a new tour, the inputs have to be the respective city numbers in a list of integers! <br>
    <li> tour_length calculates the length of the currently planned tour with return to the initial starting point. <br>
    <li> plot plots the city positions and the current tour.  
        </ul>

```python
import numpy as np
import matplotlib.pyplot as plt
import random
class traveling_salesman():
    def __init__(self, num_cities=50):
        # initialize the position of the cities
        random.seed(42)
        gridsize=100
        self.num_cities = num_cities
        self.cities = [random.sample(range(gridsize), 2) for x in range(self.num_cities)]

    def random_tour(self, num_stops=6):
        # take a random tour of a given length
        self.num_stops = num_stops
        self.tour = random.sample(range(self.num_cities), self.num_stops)
        return self.tour
        
    def new_tour(self, tour):
        # define a new tour
        tour = [int(_) for _ in tour]
        self.num_stops = len(tour)
        self.tour = tour
    
    def tour_length(self):
        # calculate tour length
        visited_cities = [self.cities[_] for _ in self.tour]
        for city in visited_cities:
            if city == visited_cities[0]:
                #start tour at the start
                traveled_distance = 0
                location = city
            else:
                # each travelled leg
                traveled_distance += np.sqrt((location[0] - city[0])**2 + (location[1] - city[1])**2)
                location = city
        # return to start
        traveled_distance += np.sqrt((visited_cities[0][0] - city[0])**2 + (visited_cities[0][1] - city[1])**2)
        return traveled_distance    

    def plot(self, tour=None):
        #plot the tour if you want to
        if tour is not None:
            self.tour = tour
        plt.plot([self.cities[_][0] for _ in range(self.num_cities)],
                 [self.cities[_][1] for _ in range(self.num_cities)],
                 linestyle='',
                 marker='*',
                markersize=11.0,
                color=[1,0,0])
        plt.plot([self.cities[self.tour[_ % self.num_stops]][0] for _ in range(self.num_cities)],
                 [self.cities[self.tour[_ % self.num_stops]][1] for _ in range(self.num_cities)],
                 'xb-')
        plt.show()
```

*Concise Explanation:*

*Exhaustive search is a method often used in programming, which enumerates and verifies many possible solutions in a certain order when the law of solving the problem cannot be found, and then finds out which candidate solutions meet the requirements as the solutions of the problem. Exhaustive search is used to find the global optimal solution, so the termination criteria is computing all cases.*

*Tuning process*

*Cpmputing all the combinations and permutations of six cities in fifty cities is tremendous, and it will spend a lot of time to running. So I think I can compute the combinations' variance to filter a lot of cases, which will decrease so many amount of computing.*

```python
# calculate shortest path function
import itertools as iter
import math

def shortest_length(combination, salesman):
    
    permutation_list = list(iter.permutations(combination, len(combination)))
    length = len(permutation_list)
    
    salesman.new_tour(tour = permutation_list[0])
    shortest_length = salesman.tour_length()
    
    travel_order = []
    city_position = []
    
    _city_list = salesman.__dict__['cities']
    
    for i in range(length):
        salesman.new_tour(tour = permutation_list[i])
        
        if salesman.tour_length() < shortest_length:
            shortest_length = salesman.tour_length()
            travel_order = permutation_list[i]
    
    salesman.new_tour(tour = travel_order)
    
    for i in travel_order:
        city_position.append(_city_list[i])
    
    return [travel_order, city_position, shortest_length, salesman]
```

```python
# calculate variance
def variance(order, saleman):
    
    x_avr = 0
    y_avr = 0
    sum_x = 0
    sum_y = 0
    position = []
    
    _cities = salesman.__dict__['cities']
    
    for _ in order:
        position.append(_cities[_])
    
    for _ in range(6):
        x_avr = x_avr + position[_][0]
        y_avr = y_avr + position[_][1]
    
    x_avr = x_avr / 6
    y_avr = y_avr / 6
    
    for _ in range(6):
        
        sum_x = sum_x + (position[_][0] - x_avr)**2
        sum_y = sum_y + (position[_][1] - y_avr)**2
    
    result = [sum_x / 6, sum_y / 6]
    
    return result
```

```python
# exhaustive search
import itertools as iter
import time

def exhaustive_search(nums):
    
    start =time.time()

    candidate_index = []

    salesman_T1Q1 = traveling_salesman(num_cities = nums)

    city_index_list = range(nums)

    temp = list(iter.combinations(city_index_list, 6))

    iteration_counter = 0

    for _ in range(len(temp)):
    
        v = variance(temp[_], salesman_T1Q1)
        if v[0] < 120 and v[1] < 120:
            candidate_index.append(temp[_])
    
        iteration_counter += 1
    
    length = len(candidate_index)


    initial = shortest_length(candidate_index[0], salesman_T1Q1)
    travel_order = initial[0]
    city_position = initial[1]
    the_shortest_length = initial[2]

    
    for i in range(length):
    
        candidate = candidate_index[i]
        result = shortest_length(candidate, salesman_T1Q1)
    
        iteration_counter += 1
    
        if result[2] < the_shortest_length:
        
            travel_order = result[0]
            city_position = result[1]
            the_shortest_length = result[2]

        
    salesman_T1Q1.new_tour(tour = travel_order)
    salesman_T1Q1.plot()
    print("The shortest length is:{}\nThe travel order is:{}\nThe city position is:{}".format(the_shortest_length,travel_order,city_position))
    print("Iteration times:{}".format(iteration_counter))
    end = time.time()
    print('Running time: %s Seconds'%(end-start))
```

*Concise Explanation*

*Random search refers to a search mode in which the search trajectory is randomly and uniformly distributed in the target distribution area under the condition that the target position basically obeys uniform distribution. Use random_tour method to generate random tour, compare to the next and find the shorter one. Because we need to find the solution within the best 5%, list every permutation and find the best 5% is impossible. And the numbers of all the cases is huge, so we can approximately view the distribution of length as a linear distribution, using random search to find the longest lenth of 6 city, combine the shortest lenth got in T1Q1, compute the 5% threshold. So the termination criteria is the length smaller than the threshold.*

```python
# random search
def random_search(iter_times, nums):
    
    start =time.time()
    
    salesman_T1Q2 = traveling_salesman(num_cities = nums)
    random_initial_tour = salesman_T1Q2.random_tour(num_stops=6)
    shortest_length = salesman_T1Q2.tour_length()
    
    travel_order = []
    city_position = []
    
    iteration_counter = 1
    
    while (iteration_counter < iter_times):
        
        random_initial_tour = salesman_T1Q2.random_tour(num_stops=6)
        length = salesman_T1Q2.tour_length()
        
        
        
        if length < shortest_length:
            
            shortest_length = length
            travel_order = random_initial_tour
            
        if shortest_length < 85.0:
            break
            
        iteration_counter += 1
    
    city_list = salesman_T1Q2.__dict__['cities']
    
    for _ in range(6):
        city_position.append(city_list[_])

    salesman_T1Q2.new_tour(tour = travel_order)
    salesman_T1Q2.plot()
    
    print("The shortest length isï¼š{}\nThe travel order is:{}\nThe city position is:{}".format(shortest_length, travel_order,city_position))
    print("The iteration times is:{}".format(iteration_counter))
    
    end = time.time()
    print('Running time: %s Seconds'%(end-start))
    
```

```python
# Genetic Algorithm

# Genetic algorithm borrows from Darwinian evolution theory to find the optimal solution. 
# It simulates the DNA sequence of organisms through encoding individual, and obtains new offspring through 
# crossover and mutation. The fitness function is used to find the fitness of each individual in the population,
# so as to eliminate the individuals with lower fitness in the selection process.


# global variable
GA_POP_SIZE
GA_CROSS_RATE
GA_MUTATION_RATE
GA_GENERATION_N

# Generating population, use two dimension array to store it
function Generate_population(number_of_individuals):
    
    pop <-- matrix[number_of_individuals, 6]
    
    # Generating multiple cities combination randomly
    for _ in number_of_individuals:
        pop[_] <-- salesman.random_tour(num_stops=6)
    
    return pop

# calculate the fitness of each individual
function Get_fitness(pop):
    
    fitness = []
    temp = []
    for _ in range(len(pop)):
        # use length of path as the fitness value
        salesman.new_tour(tour=pop[_])
        temp.append(salesman.tour_length)
    
    # the shorter path, the larger fitness, add a small value to avoid zero fitness
    fitness <-- np.max(temp) - temp + 1e-3
    return fitness

# make a new population according to the fitness, Individuals with higher fitness will appear more    
function select(fitness,pop):
    
    # select individual occurring probability, higher fitness means higher occurring probability
    index <-- np.random.choice(np.arange(POP_SIZE), size=POP_SIZE, replace=True, p=(fitness)/(fitness.sum()))
    
    return pop[index]

# crossover and mutation will make offspring, there are two hyperparameters, cross rate and mutation rate,
# a big mutation rate will lead to a unstable result, so ususally set it a small value.
function cross_and_mutation(pop, CROSS_RATE = 0.8):
    new_pop <-- np.array([])
    
    for father in pop:
        # First assign the father's value to child
        child <-- father
        
        # The crossover happen with rate
        if np.random.rand() < CROSS_RATE:
            
            # Select mother randomly
            mother <-- pop[np.random.randint(POP_SIZE)]
            # Select cross point randomly
            cross_point <-- np.random.randint(0,6)
            # The child's value after the cross point is assigned to the mother's value after the cross point
            child[cross_point:] <-- mother[cross_point:]
            
        # Every child have probability to mutation
        mutation(child)
        new_pop.append(child)
        
    return new_pop
        
# mutation function
function mutation(child, MUTATION_RATE):
    
    # mutation rate represent the probability of individual mutant, a higher mutation rate will lead to unstable
    # situation, sometimes it can even change the trend of offspring, going in a bad direction, appropriate mutation
    # rate can introduce individuas with diversity, improve offspring
    if np.random.rand() < MUTATION_RATE:
        
        # find two random mutation point
        index_1 <-- np.random.randint(0,6)
        index_2 <-- np.random.randint(0,6)
        
        # make sure they are different
        while index_1 != index_2:
            index_2 <-- np.random.randint(0,6)
        
        # switch the value on the mutation point
        temp <-- child[index_1]
        child[index_1] <-- child[index_2]
        child[index_2] <-- temp
        

# Calculate the shortest path of each population, store them in a list        
function get_shortest(pop):
    
    shortest_length = []
    salesman.new_tour(pop[0])
    shortest <-- salesman.tour_length
    for _ in range(len(pop)):
        salesman.new_tour(tour=pop[_])
        length = salesman.tour_length
        
        if length < shortest:
            shortest <-- length            
        
    shortest_length.append(shortest)
    
# There are three hyperparameters, POP_SIZE represents the size of population
# CROSS_RATE and MUTATION_RATE represent the probability of crossover and mutation
# N_GENERATION represents the times of interation
POP_SIZE <-- 500    
CROSS_RATE <-- 0.8
MUTATION_RATE <-- 0.005
N_GENERATION <-- 100

# Initialize the population
pop <-- Generate_population(POP_SIZE)

for _ in range(N_GENERATION):
    
    # get fitness of the population
    fitness <-- Get_fitness(pop)
    
    # select fitness, the individual which has higher fitness will have a higher probability of occuring
    pop <-- select(fitness,pop)
    
    # crossover and mutate the population, make new offspring
    pop <-- cross_and_mutation(pop, CROSS_RATE)
    
    # calculate the shortest path
    get_shortest(pop)
    

# Pre_implementation assessment:
# Compare to random search and exhaustive search, generic algrithm is more intelligent, it has self-organization,
# self-adaptation and self-learning habits. And it start with multiple individuals, which will avoid local optimization.
# It can find the optimal solution in less iteration times and less time
```

```python
# Your Code for T2Q2 Here

GA_POP_SIZE = 850
GA_CROSS_RATE = 0.88
GA_MUTATION_RATE = 0.05
GA_GENERATION_N = 500
INTERATION_GA = 0

def Generate_pop_GA(size):
    
    global GA_POP_SIZE, INTERATION_GA
    
    salesman_T2Q1 = traveling_salesman(num_cities=50)   
    pop = np.zeros((GA_POP_SIZE, 6), dtype = int)
    
    for _ in range(GA_POP_SIZE):
        pop[_] = salesman_T2Q1.random_tour(num_stops=6)
        INTERATION_GA += 1
        
    return [pop, salesman_T2Q1]

def Get_fitness_GA(pop, salesman):
    
    global GA_POP_SIZE, INTERATION_GA
    
    temp = np.zeros(GA_POP_SIZE)
    
    for _ in range(GA_POP_SIZE):
        
        salesman.new_tour(tour = pop[_])
        temp[_] = salesman.tour_length()
        INTERATION_GA += 1
     
    fitness = (max(temp) - temp) + 1e-4
    
    return fitness
    
def Select_GA(fitness, pop):
    
    global GA_POP_SIZE
    
    index = np.random.choice(np.arange(GA_POP_SIZE), size=GA_POP_SIZE, replace=True, p=(fitness)/(fitness.sum()))
    
    return pop[index]

def Cross_and_mutation_GA(pop):
    
    global GA_POP_SIZE, GA_CROSS_RATE, INTERATION_GA
    
    temp_pop = []
    
    for father in pop:
        child = father
        
        if np.random.rand() <= GA_CROSS_RATE:
            
            while 1:
                mother = pop[np.random.randint(GA_POP_SIZE)]
                cross_point = np.random.randint(6)
                child[cross_point:] = mother[cross_point:]
                INTERATION_GA += 1
                
                if len(set(child)) == 6:
                    break
         
        child = Mutation_GA(child)
        temp_pop.append(child)
        
    new_pop = np.asarray(temp_pop)
    
    return new_pop        
            
            
def Mutation_GA(child):
    
    global GA_MUTATION_RATE, INTERATION_GA
    
    if np.random.rand() < GA_MUTATION_RATE:
        
        while 1:
            mutation_index_1 = np.random.randint(0, 6)
            mutation_index_2 = np.random.randint(0, 6)
            
            INTERATION_GA += 1
            
            if mutation_index_1 != mutation_index_2:
                break
                
        temp = child[mutation_index_1]
        child[mutation_index_1] = child[mutation_index_2]
        child[mutation_index_2] = temp        
        return child
    
    else:     
        return child
        

def Get_shortest_GA(pop, salesman, lis, order_lis):
    
    global INTERATION_GA
    
    local_shortest = []
    
    salesman.new_tour(tour = pop[0])
    local_shortest = salesman.tour_length()
    order = pop[0]
    
    for individual in pop:
        
        salesman.new_tour(tour = individual)
        length = salesman.tour_length()
        
        if length < local_shortest:
            
            local_shortest = length
            order = individual
            
        INTERATION_GA += 1
        
    order_lis.append(order)    
    lis.append(local_shortest)

    
def Generic_Algorithm():
    
    global INTERATION_GA
    
    INTERATION_GA = 0
    
    start =time.time()
    
    pop = Generate_pop_GA(GA_POP_SIZE)[0]
    salesman_T2Q2 = Generate_pop_GA(GA_POP_SIZE)[1]
    shortest_length = []
    order_list = []
    generation_counter = 0
    
    for _ in range(GA_GENERATION_N):
    
        fitness = Get_fitness_GA(pop, salesman_T2Q2)
    
        selected_pop = Select_GA(fitness, pop)
    
        new_pop = Cross_and_mutation_GA(selected_pop)
    
        Get_shortest_GA(new_pop, salesman_T2Q2, shortest_length, order_list)
    
        pop = new_pop
        
        generation_counter += 1
        
        INTERATION_GA += 1
        
        # terminate iteration when the solution within 5% and tends to be stable
        if _ > 10:
            
            if (abs(shortest_length[_] - shortest_length[_-1])+abs(shortest_length[_-1] - shortest_length[_-2])\
            +abs(shortest_length[_-2] - shortest_length[_-3]))/3 < 1e-5:
                
                if shortest_length[-1] < 84:
                    break
    
    salesman_T2Q2.new_tour(tour = order_list[-1])
    salesman_T2Q2.plot()
    
    print("The shortest path is: ", shortest_length[-1])
    print("The travel order is:", order_list[-1])
    print("The iteration times is : ", INTERATION_GA)
    
    plt.figure(figsize=(14,7))
    plt.title("Generic Algorithm with termination")
    plt.xlabel("Generation")
    plt.ylabel("Shortest path")
    plt.plot(np.arange(generation_counter), shortest_length)
    plt.show()

    end = time.time()
    print('Running time: %s Seconds'%(end-start))

    
def Generic_Algorithm_without_termination():
    
    global INTERATION_GA
    
    INTERATION_GA = 0
    
    start =time.time()
    
    pop = Generate_pop_GA(GA_POP_SIZE)[0]
    salesman_T2Q2 = Generate_pop_GA(GA_POP_SIZE)[1]
    shortest_length = []
    order_list = []
    
    for _ in range(GA_GENERATION_N):
    
        fitness = Get_fitness_GA(pop, salesman_T2Q2)
    
        selected_pop = Select_GA(fitness, pop)
    
        new_pop = Cross_and_mutation_GA(selected_pop)
    
        Get_shortest_GA(new_pop, salesman_T2Q2, shortest_length, order_list)
    
        pop = new_pop
        
        INTERATION_GA += 1
        
    salesman_T2Q2.new_tour(tour = order_list[-1])
    salesman_T2Q2.plot()
    
    print("The shortest path is: ", shortest_length[-1])
    print("The travel order is:", order_list[-1])
    print("The iteration times is : ", INTERATION_GA)
    
    plt.figure(figsize=(14,7))
    plt.title("Generic Algorithm without termination")
    plt.xlabel("Generation")
    plt.ylabel("Shortest path")
    plt.plot(np.arange(GA_GENERATION_N), shortest_length)
    plt.show()

    end = time.time()
    print('Running time: %s Seconds'%(end-start))
     
```
